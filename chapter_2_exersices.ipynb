```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# College data set exploration

```python
college = pd.read_csv('./College.csv')
```

```python
college.head()
```

```python
college = college.rename(columns = {'Unnamed: 0' : 'College'})
college = college.set_index('College')
```

```python
college.describe()
```

```python
variables = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[variables])
plt.show()
```

```python
college.boxplot(column = ['Outstate'], by = ['Private'])
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
                          bins=[0,50,100],
                          labels = ['No', 'Yes'])
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot(column = ['Outstate'], by = ['Elite'])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
college['Top25perc'].plot.hist(ax = axs[0, 0])
college['Top10perc'].plot.hist(ax = axs[0, 1])
college['Accept'].plot.hist(ax = axs[1, 0])
college['Enroll'].plot.hist(ax = axs[1, 1])
```

```python
college.plot.scatter(x = 'Expend', y = 'Top10perc')
plt.show()
```
